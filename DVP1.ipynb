Histogram

```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/tips.csv")
plt.hist(data['tip'])
plt.title("Histogram")
plt.show()
```

Bar Chart

```python
import matplotlib.pyplot as plt
plt.bar(data['sex'], data['tip'])
plt.title("Bar Chart")
plt.xlabel('Sex')
plt.ylabel('Tip')
plt.show()
```

Scatter Plot


```python
import matplotlib.pyplot as plt
plt.scatter(data['sex'], data['tip'])
plt.title("Scatter Plot")
plt.xlabel('sex')
plt.ylabel('tip')
plt.show()
```

Box Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/tips.csv")
sns.boxplot(x='day', y='tip', data=data, hue='sex')
plt.show()
```

Barplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/tips.csv")
sns.barplot(x='day', y='tip', data=data, hue='sex')
plt.show()
```
